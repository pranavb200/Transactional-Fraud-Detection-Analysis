```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\veera\Downloads\PRANAV B\docs\projects\Transactional Fraud Detection Analysis\data\creditcard.csv")
```

```python
print("Shape of data:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum().sum())
```

```python
df.head()
```

```python
df.describe().T
#Amount — transaction values
#Class — 1 = Fraud, 0 = Legit
```

```python
fraud_count = df['Class'].value_counts()
print(fraud_count)

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_count.index, y=fraud_count.values)
plt.title("Class Distribution (0=Legit, 1=Fraud)")
plt.show()

fraud_percentage = (fraud_count[1] / fraud_count.sum()) * 100
print(f"Fraud Percentage: {fraud_percentage:.4f}%")
```

```python
# Check for NaNs again
print(df.isnull().sum().sum())

# Quick outlier visualization for 'Amount'
plt.figure(figsize=(10,4))
sns.boxplot(x=df['Amount'])
plt.title("Transaction Amount Distribution")
plt.show()
```
